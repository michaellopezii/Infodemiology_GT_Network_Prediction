```python
import numpy as np
import pandas as pd

print("Numpy version:", np.__version__)
print("Pandas version:", pd.__version__)

try:
    pickle_file = "DOH_COVID_Data_NCR_DateRegion_Filtered.pkl"  # Replace with your pickle file name
    csv_file = pickle_file.replace('.pkl', '.csv')
    
    print(f"\nReading {pickle_file}...")
    df = pd.read_pickle(pickle_file)
    
    print(f"Converting to {csv_file}...")
    df.to_csv(csv_file, index=False)
    
    print(f"\nSuccessfully converted {pickle_file} to {csv_file}")
    print(f"DataFrame shape: {df.shape}")
    
except Exception as e:
    print(f"\nError occurred: {str(e)}")
    
    # Try alternative loading method if error occurs
    try:
        print("\nTrying alternative method...")
        import pickle
        
        with open(pickle_file, 'rb') as f:
            data = pickle.load(f)
            
        if not isinstance(data, pd.DataFrame):
            data = pd.DataFrame(data)
            
        data.to_csv(csv_file, index=False)
        print(f"\nSuccessfully converted using alternative method!")
        print(f"DataFrame shape: {data.shape}")
        
    except Exception as e2:
        print(f"\nAlternative method also failed: {str(e2)}")
```
